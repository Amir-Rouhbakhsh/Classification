```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load the CSV file
df = pd.read_csv("training_metrics.csv")

# Extract relevant data
epochs = df["Epoch"]
train_acc = df["Train Accuracy"]
val_acc = df["Val Accuracy"]

# Compute standard deviation for the shaded region (simulating variations)
train_std = np.random.uniform(0.5, 1.5, len(train_acc))
val_std = np.random.uniform(0.5, 1.5, len(val_acc))

# Plot the data with a similar design
plt.figure(figsize=(10, 6), facecolor="whitesmoke")
ax = plt.gca()
ax.set_facecolor("whitesmoke")

plt.plot(epochs, train_acc, label="Train Accuracy", color="red", marker="o", linestyle="-")
plt.fill_between(epochs, train_acc - train_std, train_acc + train_std, color="red", alpha=0.3)

plt.plot(epochs, val_acc, label="Validation Accuracy", color="blue", marker="o", linestyle="-")
plt.fill_between(epochs, val_acc - val_std, val_acc + val_std, color="blue", alpha=0.3)

# Styling similar to the reference image
plt.xlabel("Epochs")
plt.ylabel("Accuracy (%)")
plt.title("Train vs Validation Accuracy Over Epochs")
plt.legend()
plt.grid(True, linestyle="--", alpha=0.6)

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load the CSV file
df = pd.read_csv("training_metrics.csv")

# Extract relevant data
epochs = df["Epoch"]
train_acc = df["Train Accuracy"]
val_acc = df["Val Accuracy"]

# Compute standard deviation for the shaded region (simulating variations)
train_std = np.random.uniform(0.5, 1.5, len(train_acc))
val_std = np.random.uniform(0.5, 1.5, len(val_acc))

# Create figure with adjusted background
fig, ax = plt.subplots(figsize=(10, 6))
fig.patch.set_facecolor("#F5F5F5")  # Light gray background
ax.set_facecolor("#EAEAF2")  # Slightly different shade for the plot area

# Plot accuracy curves
ax.plot(epochs, train_acc, label="Train Accuracy", color="red", marker="o", linestyle="-")
ax.fill_between(epochs, train_acc - train_std, train_acc + train_std, color="red", alpha=0.3)

ax.plot(epochs, val_acc, label="Validation Accuracy", color="blue", marker="o", linestyle="-")
ax.fill_between(epochs, val_acc - val_std, val_acc + val_std, color="blue", alpha=0.3)

# Customize grid and labels
ax.set_xlabel("Epochs",fontsize=16)
ax.set_ylabel("Accuracy (%)",fontsize=14)
ax.set_title("BEiT",fontsize=16)
ax.legend()
ax.grid(True, linestyle="--", alpha=0.5, color="gray")
# Add legend with increased font size
ax.legend(loc='lower right', fontsize=16)  # Adjust fontsize here (e.g., 14 or higher)
# Show the plot
plt.show()
```
